```python
- 중앙값
- IQR(Inter Quartile Range): 제 3사분위수- 제 1사분위수
- lower whisker : Q1 - 1.5 x IQR 이내인 값 중에서 최솟값 #25%
- upper whisker : Q3 + 1.5 x IQR 이내인 값 중에서 최댓값 #75%
- 이상치 : lower whisker, upper whisker 를 벗어나는 값

iqr = 9000-1000
```

```python
u_whisker_check = 9000 + 1.5*8000 
print(u_whisker_check)
print('이상치 : {}'.format(df.Wage[df.Wage>u_whisker_check]))
```

```python
iqr = 9000-1000
uw_check = 9000+1.5*iqr
print(uw_check)
st1 = np.sort(df.Wage)[::-1]
print('이상치: {}'.format(st1[st1>uw_check]))
print()
lw_check = 1000 -1.5*iqr
print(lw_check)
st2 = np.sort(df.Wage)
print('이상치: {}'.format(st2[st2<lw_check])) 
```
